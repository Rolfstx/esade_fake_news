```python
import pandas as pd
import numpy as np
from youtube_transcript_api import YouTubeTranscriptApi
from tqdm import tqdm
```

```python
df = pd.read_csv("~/Documents/GitHub/esade_fake_news/1_Climate_Change/youtube_query/data/nlp/20200507-033205_climate_change_nlp.csv", index_col=0)
df.head()
```

```python
tttt = []

with tqdm(total=len(df['id'])) as pbar:  
    for idx in df['id']:
        try:
            t = YouTubeTranscriptApi.get_transcript(idx)
        except:
            tttt.append([idx, np.nan])
            continue
        tt = [i["text"] for i in t]
        ttt = [' '.join(tt)]
        tttt.append([idx, ttt])
        pbar.update(1)

df_sub = pd.DataFrame(tttt, columns=['video_id', 'subtitles'])
df_sub.head()
```

```python
df_sub.isna().sum()
```

```python
df_sub[df_sub['subtitles'].isna()]
```

```python
df2 = df.merge(df_sub, left_on='id', right_on='video_id')
```

```python
#i checked a few nan subtitles and they don't have subtitles. 
df2[df2['subtitles'].isna()].head()
```

```python
df2.drop('video_id', inplace=True, axis=1)
```

```python
df2.to_csv("~/Documents/GitHub/esade_fake_news/1_Climate_Change/youtube_query/data/nlp/20200507-033205_climate_change_nlp_subs.csv")
```
