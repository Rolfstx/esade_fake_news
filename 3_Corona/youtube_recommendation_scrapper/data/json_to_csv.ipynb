```python
# convert json to csv
import pandas as pd

df = pd.read_json ('C:/Users/asf/Documents/GitHub/esade_fake_news/3_Corona/youtube_recommendation_scrapper/data/video-infos-coronavirus_v2_s2_b7_d7-coronavirus-20200501.json')
df.head()
```

```python
# transpose DataFrame
# transpose column to rows
df_trans  = df.T
df_trans.head()
```

```python
# drop index and create new index
df_unique = df_trans.reset_index(drop=True)
df_unique.head(10)
```

```python
#expand keys into separate rows
#cannot use recommendations column since it is a list
#gives error unhashable type
names = ['pubdate', 'views', 'dislikes', 'likes', 'duration', 'id', 'mult', 'title', 'nb_recommendations', 'depth', 'channel']

df_long = (df_unique.
           set_index(names).
           key.apply(pd.Series).stack().
           reset_index(names, 
                       name='key')
          )

df_long.head(20)
```

```python
df_long.sort_values(by='key', inplace=True)
```

```python
#This Dataframe basically shows the iteration in which the video has been find (DEPTH variable of algorithm)
# How do we interpret that? (how do i know which one it is about)
df_long.head(100)
```
