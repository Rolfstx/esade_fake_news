# Json to CSV and TOP 25

### Coronavirus is a Hoax Search:

```python
import json
import pandas as pd

with open("C:/Users/asf/Documents/GitHub/esade_fake_news/3_Corona/mikkel's scrapper/data/videos/20200614-042606_George_Floyd.json") as json_file:
    data = json.load(json_file)
```

```python
ids = [list(d.keys())[0] for d in data]
flat_data = []
for unique_id in ids:
    flat = dict(data[ids.index(unique_id)][unique_id])
    flat['id'] = unique_id
    flat['recommendations'] = len(flat['recommendations'])
    flat_data.append(flat)
flat_data = pd.DataFrame(flat_data)
```

```python
total_vids = len(flat_data)
total_vids
```

```python
flat_data.head(10)
```

```python
#write csv with full data:
flat_data.to_csv("C:/Users/asf/Documents/GitHub/esade_fake_news/3_Corona/mikkel's scrapper/data/16_06_2020_Geroge_Floyd_Full.csv", header=True)
```

```python
top25 = flat_data[['title','id','channel','genre','key']].groupby(['title','channel','genre','id']).count().sort_values(by='key',ascending=False).head(25)
```

```python
# export Data.Frame
top25.to_csv("C:/Users/asf/Documents/GitHub/esade_fake_news/3_Corona/mikkel's scrapper/data/top25/20200503-184010_coronavirus_is_a_haux_top25.csv", header=True)
```
